# Exploratory Data Analysis (EDA)

## 1. Distribution of view counts

```python
plt.figure(figsize=(10, 6))          # Creates a new figure with a specified size (width 10 inches, height 6 inches).
sns.histplot(df['view_count'], bins=30, kde=True) #Plots a histogram of the view_count column with 30 bins, and adds a Kernel Density Estimate (KDE) curve to visualize the distribution's shape.
plt.title('Distribution of View Counts') #Adds a title to the plot.
plt.xlabel('View Count')        #Labels the x-axis as "View Count".
plt.ylabel('Frequency')         #Labels the y-axis as "Frequency"
plt.show()
```

## 2. Scatter plot between view count and like count


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='view_count', y='like_count', data=df)  #Creates a scatter plot with view_count on the x-axis and 
                                                          #like_count on the y-axis. Each point on the plot represents a video.
plt.title('Scatter Plot: View Count vs Like Count')       #Adds a title to the plot.
plt.xlabel('View Count') #Labels axes accordingly
plt.ylabel('Like Count')
plt.show()
```

## 3. Box plot of view counts by category

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='category_id', y='view_count', data=df)   #Creates a box plot to show the distribution of view_count across different category_ids. 
                                                        #Each box represents the distribution of view counts within a specific category.
plt.title('Box Plot of View Counts by Category')        #Adds a title to the plot.
plt.xlabel('Category ID')                               #Labels the Axes Accordingly
plt.ylabel('View Count')
plt.show()
```

```python
# Selecting only numerical columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])   #This line selects only the columns in df that have numerical data types (float64 and int64),
                                                              #which are suitable for correlation analysis. The result is stored in numeric_df.
```

## 4. Correlation heatmap of numerical features

```python
plt.figure(figsize=(12, 8))
corr = numeric_df.corr() # Computes the pairwise correlation of all numerical columns and The correlation matrix is stored in corr.
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f') #Ensures that the correlation coefficients are displayed on the heatmap.
                                                          #fmt='.2f': Formats the numbers to show two decimal places.
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
